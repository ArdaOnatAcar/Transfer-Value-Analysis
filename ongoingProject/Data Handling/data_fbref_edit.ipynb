Imports

```python
import pandas as pd
import numpy as np
import os
```

Load and clean the FBREF data

```python
# Remove no value rows in the FBREF_20_21 data
df_fbref = pd.read_excel('../Datas/FBREF_20_21.xlsx')
df_fbref.dropna(subset=['MP (20/21)'], inplace=True)
df_fbref.reset_index(drop=True, inplace=True)  # Reset the index after dropping rows
# df_fbref.drop(columns=['Unnamed: 0'], inplace=True)  # Drop the unnamed column if it exists
print("FBREF_20_21 data cleaned.")

# Display the first few rows to verify the cleaning
df_fbref.head()
```

Save

```python
# Save the cleaned data to a new excel file
output_file = '../Datas/FBREF_20_21.xlsx'
if os.path.exists(output_file):
    os.remove(output_file)  # Remove the file if it already exists
df_fbref.to_excel(output_file, index=False)
print(f"Cleaned FBREF data saved to {output_file}.")
```
